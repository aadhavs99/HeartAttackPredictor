```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Heart Attack.csv")
```

```python
data
```

```python
data['class'] = data['class'].astype('category')
data['class'] = data['class'].cat.rename_categories({'positive': 0, 'negative': 1})
data
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split

x = data.drop(['class'], axis=1)
y = data['class']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
train_data = x_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize=(15, 8))
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
```

```python
train_data['blood_pressure'] = train_data['pressurehight']/train_data['pressurelow']
train_data['kcm_troponin_ratio'] = train_data['kcm']/train_data['troponin']
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
```

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()

forest.fit(x_train, y_train)
```

```python
forest.score(x_test, y_test)
```

```python
test_data = x_test.join(y_test)

test_data['blood_pressure'] = test_data['pressurehight']/test_data['pressurelow']
test_data['kcm_troponin_ratio'] = test_data['kcm']/test_data['troponin']

test_data['class'] = test_data['class'].astype('category')
test_data['class'] = test_data['class'].cat.rename_categories({'positive': 0, 'negative': 1})

x_test, y_test = test_data.drop(['class'], axis=1), test_data['class']
test_data
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()

forest.fit(x_train_scaled, y_train)
```

```python
forest.score(x_test_scaled, y_test)
```

```python
from sklearn.model_selection import GridSearchCV

forest = RandomForestRegressor()

param_grid = {
    "n_estimators": [3, 10, 30],
    "max_features": [2, 4, 6, 8]
}

grid_search = GridSearchCV(forest, param_grid, cv=5, scoring="neg_mean_squared_error", return_train_score=True)

grid_search.fit(x_train_scaled, y_train)
```

```python
grid_search.best_estimator_
```
